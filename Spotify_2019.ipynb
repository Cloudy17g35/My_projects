## Data visuaalization

```python
plt.figure(figsize=(10,10))
plt.title("Correlation map")
ax = sns.heatmap(data.corr(),
                linewidth=3.1,
                annot=True,
                center=1)
```

Now we re gonna make a countplot by genre:

```python
sns.catplot(y = "Genre", kind = "count",
           palette = "pastel", edgecolor = ".6",
           data = data)
plt.title("Genre count", fontsize=16)
plt.show()
```

Visualising beats per minute for different genre's

```python
plt.figure(figsize= (6,6))
sns.barplot(x = data["Genre"], y = data["Beats.Per.Minute"], palette = "ocean")
plt.xticks(rotation = 90)
plt.title("Beats per Genre", fontsize=18)
plt.show()
```

We can see the biggest beats per minute for trap

Visualising Liveness of songs produced by different Artists

```python
plt.figure(figsize=(6,6))
ax=plt.axes()
ax.scatter(data['Artist.Name'], data['Liveness'])
ax.set(xlabel='Artist Name',
       ylabel='Liveness',
       title='Liveness in Artits songs');
plt.xticks(rotation=90)
plt.show()
```
